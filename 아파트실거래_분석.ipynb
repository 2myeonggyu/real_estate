```python
# api를 통해 넘겨 받은 데이터를 테이블 형태로 분석 할 수 있게 만드는 함수(get_item)와 결과

import xml.etree.ElementTree as ET

def get_items(response):
    root = ET.fromstring(response.content)
    item_list = []
    for child in root.find('body').find('items'):
        elements = child.findall('*')
        data = {}
        for element in elements:
            tag = element.tag.strip()
            text = element.text.strip()
            # print tag, text
            data[tag] = text
        item_list.append(data)  
    return item_list

# 함수 get_item() 을 이용하여 얻은 정보를 테이블화 
items_list = get_items(res)
items = pd.DataFrame(items_list) 
items.head()
```

## 3. '노원구' 거래 분석
- 기준일자 : ~2021.12월
- 지역 코드(법정동 코드) : 11350

```python
# 기준 일자 설정

year = [str("%02d" %(y)) for y in range(2006, 2022)]
month = [str("%02d" %(m)) for m in range(1, 13)]
base_date_list = ["%s%s" %(y, m) for y in year for m in month ]

# 지역 코드 설정
gu_code = '11350'
```

```python
print(base_date_list[:3])
print(base_date_list[-3:])
print(gu_code)
```

```python
# 데이터 요청 함수
def get_data(dong_code, base_date , url=url, service_key=service_key):
    url = url
    service_key = service_key   
    payload = "LAWD_CD=" + dong_code + "&" + \
              "DEAL_YMD=" + base_date + "&" + \
              "serviceKey=" + service_key + "&"

    res = requests.get(url + payload)
    
    return res
```

```python
# 위의 요청 함수 이용하여, 기준 일자에 포함되는 모든 거래 데이터 끌어오기
items_list = []
for base_date in tqdm(base_date_list):
    try:
        res = get_data(dong_code = gu_code, base_date = base_date)
        items_list += get_items(res)
    except:
        pass
    
print(len(items_list) )

data = pd.DataFrame(items_list) 
data.head()
```

```python
# 하계동 '현대/우성' 아파트 데이터만 활용하여 시각해보자.
data_apt1 = data.query("법정동 == '하계동' and (아파트 == '현대' or 아파트 == '우성')").copy()
```

```python
# 참고 : https://docs.datapane.com/examples-and-tutorials/interactive-filters

data_query = data_apt1.copy()

fig = go.Figure()

area_list = list(data_query['전용면적'].unique())

for area in area_list:
    fig.add_trace(
        go.Scatter(
            x = data_query['datetime'][data_query['전용면적']==area],
            y = data_query['거래금액'][data_query['전용면적']==area],
            name = area, visible = True
        )
    )
    
buttons = []

for i, area in enumerate(area_list):
    args = [False] * len(area_list)
    args[i] = True
    
    #create a button object for the country we are on
    button = dict(label = area,
                  method = "update",
                  args=[{"visible": args}])
    
    #add the button to our list of buttons
    buttons.append(button)
    
fig.update_layout(
    updatemenus=[dict(
                    active=0,
                    type="dropdown",
                    buttons=buttons,
                    x = 0,
                    y = 1.1,
                    xanchor = 'left',
                    yanchor = 'bottom'
                )], 
    autosize=False, # default:False
    width=1000,
    height=800,
    #xaxis1_rangeslider_visible = True # 아래에 x-range 정보 추가로 나타내준다. 하지만 이때 y-axis가 고정되는 문제 있는듯
)


fig
```
